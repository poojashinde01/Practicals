```python
#4 Write a program to implement POS Tagging using HMM and Neural Models.

import nltk
from nltk import tokenize, tag, chunk

# Download necessary NLTK data files
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')

# Sample paragraph
para = "Hello! My name is Beena Kapadia. Today you'll be learning NLTK."

# Sentence tokenization
sents = tokenize.sent_tokenize(para)
print("\nSentence Tokenization\n===================\n", sents)

# Word tokenization
print("\nWord Tokenization\n===================\n")
for sentence in sents:
    words = tokenize.word_tokenize(sentence)
    print(words)

# POS Tagging
print("\nPOS Tagging\n===========\n")
tagged_words = []
for sentence in sents:
    words = tokenize.word_tokenize(sentence)
    tagged_words.append(tag.pos_tag(words))
print(tagged_words)

# Chunking
print("\nChunking\n========\n")
trees = []
for tagged_sentence in tagged_words:
    trees.append(chunk.ne_chunk(tagged_sentence))
for tree in trees:
    print(tree)
```

```python
#5	Write a program to Implement syntactic parsing of a given text
import nltk
from nltk import tokenize
from nltk.tree import Tree
import pydot
from IPython.display import Image, display

def tree_to_dot(tree):
    """
    Convert NLTK Tree to DOT format for visualization with Graphviz.
    """
    def add_edges(node, graph):
        if isinstance(node, Tree):
            for child in node:
                if isinstance(child, Tree):
                    graph.add_edge(pydot.Edge(str(node), str(child)))
                    add_edges(child, graph)
                else:
                    graph.add_edge(pydot.Edge(str(node), child))
        else:
            graph.add_edge(pydot.Edge(str(node), node))

    graph = pydot.Dot(graph_type='digraph')
    add_edges(tree, graph)
    return graph.to_string()

# Define a context-free grammar (CFG)
grammar1 = nltk.CFG.fromstring("""
    S -> VP NP
    VP -> 'Book'
    NP -> Det Noun
    Det -> 'that'
    Noun -> 'flight'
""")

# Sentence to parse
sentence = "Book that flight"

# Tokenize the sentence
all_tokens = tokenize.word_tokenize(sentence)
print("Tokens:", all_tokens)

# Create a parser with the given grammar
parser = nltk.ChartParser(grammar1)

# Parse the sentence and save the parse tree as a DOT file and convert to PNG
for i, tree in enumerate(parser.parse(all_tokens)):
    print(tree)
    dot_str = tree_to_dot(tree)
    dot_file = f'tree_{i}.dot'

    # Save DOT file
    with open(dot_file, 'w') as f:
        f.write(dot_str)

    # Convert DOT file to PNG using Graphviz
    (graph,) = pydot.graph_from_dot_file(dot_file)
    png_file = f'tree_{i}.png'
    graph.write_png(png_file)
    print(f"Parse tree saved as {png_file}")

    # Display the PNG file if using Jupyter Notebook
    display(Image(filename=png_file))
```

```python
# 8 Write a program to implement text Summarization.
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
from collections import defaultdict
import re

nltk.download('stopwords')
# Sample text
text = """
Apple Inc. is planning to open a new office in San Francisco on October 12, 2024.
Tim Cook will be present at the opening ceremony.
Apple is known for its innovative technology and design.
The new office is expected to create many jobs in the area.
San Francisco is a major tech hub with many companies in the field.
"""

# Preprocess the text
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    text = re.sub(r'\W', ' ', text)  # Remove non-word characters
    text = re.sub(r'\d+', ' ', text)  # Remove numbers
    return text

# Tokenize the text into sentences
sentences = sent_tokenize(text)
preprocessed_text = preprocess_text(text)
words = word_tokenize(preprocessed_text)
stop_words = set(stopwords.words('english'))

# Create a frequency distribution of words
word_frequencies = defaultdict(int)
for word in words:
    if word not in stop_words:
        word_frequencies[word] += 1

# Calculate sentence scores
sentence_scores = defaultdict(int)
for sentence in sentences:
    sentence_word_count = len(word_tokenize(sentence))
    for word in word_tokenize(sentence.lower()):
        if word in word_frequencies:
            sentence_scores[sentence] += word_frequencies[word]

# Calculate the average score
average_score = sum(sentence_scores.values()) / len(sentence_scores)

# Create summary based on sentence scores
summary = ""
for sentence in sentences:
    if sentence in sentence_scores and sentence_scores[sentence] > (1.2 * average_score):
        summary += " " + sentence

# Print the summary
print(summary.strip())

```

```python
# 9 Consider a scenario of applying NLP in Customer Service. Design and develop an application that demonstrates NLP operations for working with tasks and data like voice calls, chats, Ticket Data, Email Data. Process the data to understand the voice of the Customer (intent mining, Top words, word cloud, classify topics). Identify issues, replace patterns and gain insight into sales chats.

import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from transformers import pipeline
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Sample data
data = {
    'source': ['chat', 'email', 'ticket', 'voice_call'],
    'text': [
        'I am having issues with my billing statement. Can you help?',
        'I need to update my shipping address for my recent order.',
        'The product I received is defective and I want a refund.',
        'Can you provide more information about the new features in the latest update?'
    ]
}

df = pd.DataFrame(data)

# Data preprocessing
nltk.download('punkt')
nltk.download('stopwords')

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    return ' '.join(tokens)

df['processed_text'] = df['text'].apply(preprocess_text)

# Intent Mining
classifier = pipeline('zero-shot-classification', model='facebook/bart-large-mnli')
candidate_labels = ['billing', 'shipping', 'refund', 'product_info', 'other']

def classify_intent(text):
    result = classifier(text, candidate_labels)
    return result['labels'][0]

df['intent'] = df['text'].apply(classify_intent)

# Top Words Extraction and Word Cloud
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['processed_text'])
word_freq = dict(zip(vectorizer.get_feature_names_out(), X.toarray().sum(axis=0)))

wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

# Topic Classification
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['processed_text'])

lda = LatentDirichletAllocation(n_components=3, random_state=42)
lda.fit(X)

def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic #{topic_idx}:")
        print(" ".join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))

display_topics(lda, vectorizer.get_feature_names_out(), 10)

# Issue Identification and Pattern Replacement
issue_counts = df['intent'].value_counts()
print("Issue Counts:")
print(issue_counts)

# Example sales chat analysis
sales_chat = "I am interested in the features of your product and how it compares with competitors."
sales_chat_processed = preprocess_text(sales_chat)
sales_chat_intent = classify_intent(sales_chat)
print(f"Sales Chat Intent: {sales_chat_intent}")
```

```python
# 10 Consider a scenario of Online Review and demonstrate the concept of sentiment analysis and emotion mining by applying various approaches like lexicon-based approach and rule-based approaches.
import pandas as pd
import re
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from textblob import TextBlob

# Sample reviews
data = {
    'review': [
        'I absolutely love this product! It works great and exceeded my expectations.',
        'Terrible experience. The product broke within a week and customer service was unhelpful.',
        'It’s okay, not the best but decent for the price.',
        'Excellent quality and fast delivery. Will definitely buy again!',
        'Very disappointed with the quality. Not worth the money.'
    ]
}

df = pd.DataFrame(data)

# Lexicon-Based Sentiment Analysis

# VADER
analyzer = SentimentIntensityAnalyzer()
def analyze_sentiment_vader(review):
    score = analyzer.polarity_scores(review)
    return score['compound']

df['sentiment_vader'] = df['review'].apply(analyze_sentiment_vader)
def get_sentiment_label(compound_score):
    if compound_score >= 0.05:
        return 'Positive'
    elif compound_score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment_label_vader'] = df['sentiment_vader'].apply(get_sentiment_label)

# TextBlob
def analyze_sentiment_textblob(review):
    blob = TextBlob(review)
    return blob.sentiment.polarity

df['sentiment_textblob'] = df['review'].apply(analyze_sentiment_textblob)
df['sentiment_label_textblob'] = df['sentiment_textblob'].apply(lambda x: get_sentiment_label(x))

# Rule-Based Sentiment Analysis
positive_words = ['love', 'great', 'excellent', 'fast', 'buy', 'happy']
negative_words = ['terrible', 'broke', 'unhelpful', 'disappointed', 'poor', 'bad']

def custom_sentiment_analysis(review):
    review = review.lower()
    positive_score = sum(word in review for word in positive_words)
    negative_score = sum(word in review for word in negative_words)

    if positive_score > negative_score:
        return 'Positive'
    elif negative_score > positive_score:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment_custom'] = df['review'].apply(custom_sentiment_analysis)

# Emotion Mining
emotion_keywords = {
    'joy': ['love', 'happy', 'great', 'pleased', 'excited'],
    'anger': ['terrible', 'broke', 'unhelpful', 'disappointed', 'poor'],
    'surprise': ['exceeded', 'amazing', 'fantastic'],
    'sadness': ['bad', 'worst', 'dissatisfied']
}

def detect_emotions(review):
    emotions_detected = []
    review = review.lower()
    for emotion, keywords in emotion_keywords.items():
        if any(keyword in review for keyword in keywords):
            emotions_detected.append(emotion)
    return ', '.join(emotions_detected) if emotions_detected else 'None'

df['emotions'] = df['review'].apply(detect_emotions)

print(df[['review', 'sentiment_label_vader', 'sentiment_label_textblob', 'sentiment_custom', 'emotions']])
```

```python
#11 Apply NLP in Banking, Financial Services, and Insurance. Design Application to detect frauds and work with SMS data


import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split

# Sample SMS data
data = {
    'sms_id': [1, 2, 3, 4, 5],
    'text': [
        'Your account has been accessed from a new device. If this was not you, please call us immediately.',
        'Congratulations! You have won $1000. Click here to claim your prize.',
        'Your loan application has been approved. Please visit our branch to complete the process.',
        'Alert: Suspicious login attempt detected. Please verify your identity.',
        'Get a free insurance quote now. Call us at 555-1234.'
    ]
}

df = pd.DataFrame(data)

# Data Preprocessing
nltk.download('punkt')
nltk.download('stopwords')

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    return ' '.join(tokens)

df['processed_text'] = df['text'].apply(preprocess_text)

# Keyword-Based Fraud Detection
fraud_keywords = ['alert', 'suspicious', 'verify', 'call', 'won', 'prize']

def detect_fraud_keywords(text):
    for keyword in fraud_keywords:
        if keyword in text:
            return 1  # Fraudulent
    return 0  # Not fraudulent

df['fraud_keyword_detection'] = df['processed_text'].apply(detect_fraud_keywords)

# Machine Learning-Based Fraud Detection
labeled_data = {
    'text': [
        'Your account has been accessed from a new device.',
        'Congratulations! You have won $1000.',
        'Your loan application has been approved.',
        'Alert: Suspicious login attempt detected.',
        'Get a free insurance quote now.'
    ],
    'label': [1, 1, 0, 1, 0]
}

df_labeled = pd.DataFrame(labeled_data)

X_train, X_test, y_train, y_test = train_test_split(df_labeled['text'], df_labeled['label'], test_size=0.2, random_state=42)

model = make_pipeline(TfidfVectorizer(), MultinomialNB())
model.fit(X_train, y_train)

df['fraud_ml_detection'] = model.predict(df['text'])

# Reporting and Alerts
def generate_alerts(fraud_df):
    for index, row in fraud_df.iterrows():
        print(f"Alert: SMS ID {row['sms_id']} is flagged as fraudulent.")

fraud_cases = df[df['fraud_ml_detection'] == 1]
print("Fraudulent SMS Messages:")
print(fraud_cases[['sms_id', 'text']])
generate_alerts(fraud_cases)
```

```python
# 12 Demonstrate the use of NLP in designing Virtual Assistants. Apply LSTM, build conversational Bots
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# Sample dataset
data = {
    'intent': ['greeting', 'goodbye', 'thanks', 'hours', 'location'],
    'patterns': [
        ['Hi', 'Hello', 'How are you?', 'What’s up?'],
        ['Bye', 'Goodbye', 'See you later'],
        ['Thank you', 'Thanks', 'Appreciate it'],
        ['What are your hours?', 'When are you open?', 'Opening time'],
        ['Where are you located?', 'Location', 'Address']
    ],
    'responses': [
        ['Hello! How can I help you today?', 'Hi there! What can I do for you?', 'Hey! How can I assist?'],
        ['Goodbye! Have a great day!', 'See you later!', 'Take care!'],
        ['You’re welcome!', 'No problem!', 'Glad I could help!'],
        ['We are open from 9 AM to 5 PM, Monday to Friday.', 'Our working hours are 9 to 5, Monday through Friday.', 'We are open 9 AM to 5 PM, weekdays.'],
        ['We are located at 123 Main Street, Cityville.', 'Our address is 123 Main Street, Cityville.', 'You can find us at 123 Main Street, Cityville.']
    ]
}

df = pd.DataFrame(data)

# Tokenization and padding
tokenizer = Tokenizer()
tokenizer.fit_on_texts([pattern for patterns in df['patterns'] for pattern in patterns])
vocab_size = len(tokenizer.word_index) + 1

# Convert patterns to sequences
X = []
for patterns in df['patterns']:
    sequences = tokenizer.texts_to_sequences(patterns)
    X.extend(sequences)

# Pad sequences
max_length = max(len(seq) for seq in X)
X_padded = pad_sequences(X, maxlen=max_length, padding='post')

# Prepare labels
le = LabelEncoder()
y = le.fit_transform(df['intent'])
y = np.array([y[i] for i, patterns in enumerate(df['patterns']) for _ in patterns])

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_padded, y, test_size=0.2, random_state=42)

# Build the model
model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=64, input_length=max_length))
model.add(LSTM(64))
model.add(Dense(len(df['intent'].unique()), activation='softmax'))

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=10, validation_split=0.2, batch_size=8)

# Function to predict intent
def predict_intent(text):
    sequence = tokenizer.texts_to_sequences([text])
    padded_sequence = pad_sequences(sequence, maxlen=max_length, padding='post')
    prediction = model.predict(padded_sequence)
    intent_index = np.argmax(prediction, axis=1)[0]
    return df['intent'].iloc[intent_index]

# Function to get response
def get_response(intent):
    responses = df[df['intent'] == intent]['responses'].values[0]
    return np.random.choice(responses)

# Example usage
user_input = 'Hi there!'
intent = predict_intent(user_input)
response = get_response(intent)

print(f"User: {user_input}")
print(f"Bot: {response}")
```
